```python
TASK 5:
class Node:
    def _init_(self, value):
        self.value = value
        self.children = []

    def add_child(self, child):
        self.children.append(child)

def dfs_stack(root):
    stack = [root]
    visited = set()

    while stack:
        node = stack.pop()
        if node.value not in visited:
            print(node.value, end=" ")  
            visited.add(node.value)
            stack.extend(reversed(node.children))  

a = Node('A')
b = Node('B')
c = Node('C')
d = Node('D')
e = Node('E')

a.add_child(b)
a.add_child(c)
b.add_child(d)
b.add_child(e)

print("DFS Traversal:")
dfs_stack(a)
```

```python
class TreeNode:
    def _init_(self, value):
        self.value = value
        self.left = None
        self.right = None

def inorder(root):
    if root:
        inorder(root.left)
        print(root.value, end=" ") 
        inorder(root.right)

def preorder(root):
    if root:
        print(root.value, end=" ") 
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.value, end=" ")  

root = TreeNode('A')
root.left = TreeNode('B')
root.right = TreeNode('C')
root.left.left = TreeNode('D')
root.left.right = TreeNode('E')

print("\nInorder Traversal:")
inorder(root)

print("\nPreorder Traversal:")
preorder(root)

print("\nPostorder Traversal:")
postorder(root)
```
